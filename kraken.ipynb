```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create fabricated data
np.random.seed(0)
num_players = 100
ages = np.random.randint(18, 40, num_players)
win_loss_rates = np.random.uniform(0, 1, num_players)

# Create a DataFrame
data = pd.DataFrame({'Age': ages, 'Win/Loss Rate': win_loss_rates})

# Calculate correlation
correlation = data['Age'].corr(data['Win/Loss Rate'])

# Plot the data
sns.scatterplot(data=data, x='Age', y='Win/Loss Rate')
plt.title(f"Corr between Kraken Player Age and Win/Loss Rate: {correlation:.2f}")
plt.xlabel("Player Age")
plt.ylabel("Win/Loss Rate")
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Create fabricated data
np.random.seed(0)
num_samples = 100
nick_fucking_around = np.random.randint(0, 10, num_samples)
work_done = 8 + 2 * nick_fucking_around + np.random.normal(0, 2, num_samples)

# Create a DataFrame
data = pd.DataFrame({'Nick fucking around': nick_fucking_around, 'Amount of work getting done': work_done})

# Perform linear regression
regression_model = LinearRegression()
X = data[['Nick fucking around']]
y = data['Amount of work getting done']
regression_model.fit(X, y)
predicted_work_done = regression_model.predict(X)

# Plot the data and regression line
sns.scatterplot(data=data, x='Nick fucking around', y='Amount of work getting done')
plt.plot(X, predicted_work_done, color='red', linewidth=2, label='Regression Line')
plt.title("Relationship between Nick fucking around and Amount of work getting done")
plt.xlabel("Nick fucking around")
plt.ylabel("Amount of work getting done")
plt.legend()
plt.show()
```
